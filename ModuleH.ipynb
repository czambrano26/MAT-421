Section 21.1: ODE Initial Value Problem Statement

---
A differential equation is a relationship between a function f(x), its independent variable x, and any number of its derivatives. 

An ordinary differential equation (ODE) is a differential equation where the independent variable (and also the derivatives), is in one dimension. 

A partial differential equation (PDE) is a general form differential equation where x is a vector containing the independent variables and the partial derivatives can be of any order and with repect to any combination of variables. 

A general solution to a differential equation is a g(x) that satisfies the differential equation.

For an ODE of order n, a particular solution is a p(x) that satisfies the differential equation and n explicitly known values of the solution, or its derivatives, at certain points. 



Section 21.2: Reduction of Order

---
Many numerical methods for solving initial value problems are designed specifically to solve first-order differential equations. To make these solvers work for higher order differnetial equations, we must reduce the order of the differential equation to first order. 

To reduce the order, consider a vector S(t), which is the state of the system as a function of time. The derivative of S(t) can be written without f(t) or its derivatives. This equation is in the form of a first-order differential equation in S. 


Section 21.3: The Euler Method

---
The Explicit Euler Formula allows us to compute an approcimation for the state at S(t_j + 1) given the state at S(t_j). We can use this formmula to integrate the states up to S(t_f). These S(t) values are an approximation for the solution of the differential equation. 

We can compute S(t_j) for every t_j in t using the following steps:
1. Store S_0 = S(t_0) in an array, S
2. Compute S(t_1) = S_0 + h*F(t_0, S_0)
3. Store S_1 = S(t_1) in S
4. Compute S(t_2) = S_1 + h*F(t_1, S_1)
5. Store S_2 = S(t_1) in S
6. ...
7. Compute S(t_f) = s_(f-1) + h*F(t_(f-1), S_(f-1))
8. Store S_f = S(t_f) in S
9. S is an approximation of the solution to the intial value problem

Ex. The differential equation df(t)/dt = -e^(-t) with initial condition f_0 = 1 has the exact solution f(t) = e^(-t). Approximate the solution to the initial value problem between 0 and 1 in increments of 0.05 using the Explicit Euler Formula. Plot the difference between the appriximated solution and the exact solution.

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
%matplotlib inline

# Define parameters
f = lambda t, s: -np.exp(-t) # ODE
h = 0.05 # Step size
t = np.arange(0, 1 + h, h) # Numerical grid
s0 = 1 # Initial Condition

# Explicit Euler Method
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (12, 8))
plt.plot(t, s, 'co--', label='Approximate')
plt.plot(t, np.exp(-t), 'r', label='Exact')
plt.title('Approximate and Exact Solution \
for Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

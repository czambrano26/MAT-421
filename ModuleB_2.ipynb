Section 17.4: Lagrange Polynomial Interpolation

---
Lagrange polynomial interpolation finds a single polynomial that goes htrough all data points.

The lagrange polynomial, L(x), should have the property L(xi)=yi for every point in the data set.

Find the Lagrange basis polynomials for the data set x = [0, 1, 2] and y = [3, 1, 5]. Plot each polynomial and verify that Pi(xj)=1 when i=j and Pi(xj)=0 when i does not equal j. 

```python
import numpy as np
import numpy.polynomial.polynomial as poly
import matplotlib.pyplot as plt

x = [0, 1, 2]
y = [3, 1, 5]

P1_coeff = [1,-1.5,.5]
P2_coeff = [0, 2,-1]
P3_coeff = [0,-.5,.5]

# get the polynomial function
P1 = poly.Polynomial(P1_coeff)
print("P1 = ", P1)
P2 = poly.Polynomial(P2_coeff)
print("P2 = ", P2)
P3 = poly.Polynomial(P3_coeff)
print("P3 = ", P3)

x_new = np.arange(-1.0, 3.1, 0.1)

fig = plt.figure(figsize = (10,8))
plt.plot(x_new, P1(x_new), 'b', label = 'P1')
plt.plot(x_new, P2(x_new), 'r', label = 'P2')
plt.plot(x_new, P3(x_new), 'g', label = 'P3')

plt.plot(x, np.ones(len(x)), 'ko', x, np.zeros(len(x)), 'ko')
plt.title('Lagrange Basis Polynomials')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()
```

Compute and plot the Lagrange polynomial.

```python
L = 3*P1 + 1*P2 + 5*P3

fig = plt.figure(figsize = (10,8))
plt.plot(x_new, L(x_new), 'b', x, y, 'ro')
plt.title('Lagrange Polynomial')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Rather than calculate everything like in the example above, we can use the lagrange function in scipy. 

Compute and plot the Lagrange polynomial and verify that it goes through each of the data points x=[3, 4, 5] and y=[2, 6, 4]

```python
from scipy.interpolate import lagrange

x_2 = [3, 4, 5]
y_2 = [2, 6, 4]

x_new = np.arange(1.5, 6.5, 0.1)

f_new = lagrange(x_2,y_2)

fig = plt.figure(figsize = (10,8))

plt.plot(x_new, f_new(x_new), 'b', x_2, y_2, 'ro')
plt.title('Lagrange Polynomial')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
